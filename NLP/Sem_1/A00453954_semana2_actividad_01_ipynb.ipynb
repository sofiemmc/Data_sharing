```python
# Dropping last row with NAN value
df=df.dropna()
```

```python
df['Date of posting']
```

```python
import re

# Pattern to match the time
pattern = r'(\d{2})(:\d{2})(:\d{2})'

# New list to store the formatted time
output_list = []
# Loop through each input string and extract the time
for string in df['Date of posting']:
    match = re.search(pattern, string)
    if match:
        # Convert the time to 12-hour format
        i=int(match.group(1))
        if i>=12:
            i=i-12
            i=str(i)
            time_12h=i + match.group(2) + match.group(3) +'pm'
            
        else:
            i=str(i)
            time_12h=i + match.group(2) + match.group(3) +'am'


        output_list.append(time_12h)

df['Hora de publicacion']=(output_list)
df.head(3)

```
